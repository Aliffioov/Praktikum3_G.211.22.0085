```python
# Input berupa sebuah daftar huruf
input_list = ['p', 'r', 'o', 'g', 'r', 'a', 'm', 'm', 'i', 'n', 'g']

# Inisialisasi string kosong untuk penyimpanan hasil
result_string = ''

# Menggunakan loop for untuk menggabungkan huruf-huruf
for letter in input_list:
    result_string += letter

# Mencetak string hasil penggabungan
print(result_string)
```
